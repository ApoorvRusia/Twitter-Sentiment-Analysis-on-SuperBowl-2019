```
#Step 5- Defining the cleaning function
import re
def clean_tweet(tweet):
  '''Utility function to clean the tweets by removing links and special characters using regex'''
  return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\s+)"," ", tweet).split())
```

```
#Step 6 - Find Related Tweets 
tweets = api.search('Super Bowl', count=10000)

# Create a clean pandas dataframe as follows:
data = pd.DataFrame(data = [clean_tweet(tweet.text) for tweet in tweets], columns=['Tweets'])

# display the top 10 tweets
display(data.head(10))

```

```
#gathering lexicon data
import nltk
nltk.download('vader_lexicon') # this lexicon Vader stands for Valence Aware Dictionary and Sentiments Reasoner which is atuned to sentiments expressed in Social media
```

```
#Step 7 - Go through the tweets to analyze their sentiments

sid = SentimentIntensityAnalyzer()

list = []
for index, row in data.iterrows():
  ss = sid.polarity_scores(row['Tweets'])
  list.append(ss)

se = pd.Series(list)
data['Polarity'] = se.values


# displaying first 10 rows of the dataframe
display(data.head(100))
```
